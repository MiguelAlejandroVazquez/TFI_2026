```python
import matplotlib.pyplot as plt
import networkx as nx


G = nx.DiGraph()

conexiones = [
    (0, 1),
    (0, 2),
    (1, 2),
    (1, 3),
    (2, 3),
    (2, 4),
    (3, 4),
    (3, 5),
    (4, 5),
    (4, 6),
    (5, 6),
    (5, 7),
    (6, 7),
    (6, 0),
    (7, 0),
    (7, 1),
]
G.add_edges_from(conexiones)

nombres_nodos = {
    0: "Nodo A",
    1: "Nodo B",
    2: "Nodo C",
    3: "Nodo D",
    4: "Nodo E",
    5: "Nodo F",
    6: "Nodo G",
    7: "Nodo H",
}

nx.draw_circular(
    G,
    labels=nombres_nodos,
    node_color="lightblue",
    node_size=1200,
    font_size=10,
)
print("--- VERIFICACIÓN DE CONEXIONES ---")

for numero, nombre in nombres_nodos.items():

    vecinos = list(G.successors(numero))


    nombre_vecino1 = nombres_nodos[vecinos[0]]
    nombre_vecino2 = nombres_nodos[vecinos[1]]

    print(f"{nombre} está conectado directamente con: {nombre_vecino1} y {nombre_vecino2}")
plt.show()
```
